```python
import pandas as pd
```

```python
df1=pd.DataFrame({"Perro":[1,2,3,4,5,6,10,2,2,2],
                  "Gato":[1,2,3,4,5,6,10,2,2,2],
                  "Conejo":[1,2,3,4,5,6,10,2,2,2]})
                  
```

```python
names=list(df1.columns)
names
```

```python
df2=pd.DataFrame(df1.loc[:,names[0]])
df2=df2.rename(columns={names[0]:"Values"})
```

```python
def concatData(data):
    names=list(data.columns)
    datas=[]
    for i in range(len(names)):
        newData=pd.DataFrame(data.loc[:,names[i]])
        newData=newData.rename(columns={names[i]:"Values"})
        newData["Name"]=names[i]
        datas.append(newData)
    dataConcat=pd.concat(datas)
    dataConcat=dataConcat.reset_index(drop=True)
    return dataConcat
    
```

```python
ListDatas=concatData(df1)
```
